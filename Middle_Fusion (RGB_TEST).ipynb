```
import os
import cv2
from google.colab.patches import cv2_imshow
%tensorflow_version 1.0
import tensorflow as tf
from tensorflow.core.protobuf import saver_pb2
import numpy
import scipy
import sys
import matplotlib.pyplot
from PIL import Image
from matplotlib import pyplot as plt
import scipy.misc
import random
from google.colab import output
import time
import numpy as np
import datetime
from openpyxl import load_workbook
import matplotlib.pyplot as plt
import csv
import pandas as pd
random.seed(5)

#############################################################################################################
#--------##########################################################################################--------------#
#-----------------####################--->>> Preparing Data <<<---####################-------------------------------#
#--------##########################################################################################--------------#
#############################################################################################################


####################--->> directories Paths <<---####################

dir_txt = "/content/drive/My Drive/GP/our_drEYEve/steering_speed_new"
dir_frames = "/content/Data_set/Frames"
dir_gaze = "/content/Data_set/Gaze"
LOGDIR = "/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Results_files/ V9'_#TEST_RGB_2 _2020-05-27 03:19:15.490475" 

####################--->> Hyper Parameters & Control <<---####################

dataset_portion = 1                           

train_test_ratio = 0.8                        

visualization = 7                             

epochs = 5

batch_size = 128

learning_rate = 0.0001*0.5*0.5              

L2NormConst = 0.0001*0.5

dropout_keep_prob = 0.85

log_flag = 1                                  


####################--->> Comment <<---####################

comment= " #...# "

name= " V9''_#TEST_RGB_3 "

####################--->> Loading Data <<---####################

loc_frames = []
loc_gaze = []
S_angle = []
speed=[]
Nvidia_data=[]

txt_files = [f for f in os.listdir(dir_txt) if not f.startswith('.')]

for i in range(len(txt_files)):
  with open(dir_txt+"/"+txt_files[i]) as f:
    for line in f:                                         #loop on each line in a specific text file

        # Saving the location of each image
        loc_frames.append(dir_frames+"/" +(txt_files[i][0:2])+"/" + line.split()[0]) # both frames and gaze have the same name of the picture .. just saving in different lists.
     
        # Saving the location of each gaze map
        loc_gaze.append(dir_gaze+"/" +(txt_files[i][0:2])+"/" + line.split()[0])
        
        # Saving the steering angles
        S_angle.append(float(line.split()[1])/527) 
        
        # Saving the speed
        speed.append(float(line.split()[2]))
        


num_images=len(speed)                                     

####################--->> randomly shuffle arrayes for better training <<---####################

loc_frames_testing = loc_frames[visualization*7470:(visualization*7470)+7469]           
S_angle_testing = S_angle[visualization*7470:(visualization*7470)+7469]                 

#===============================================ٍ Shuffling the data ================================================================

c = list(zip(loc_frames,loc_gaze, S_angle,speed))
random.shuffle(c) 
loc_frames,loc_gaze, S_angle,speed = zip(*c)
      
####################--->> Creating arrays <<---####################

xs = [] #final input frames "locations" to the network
gs = [] #Gazemaps "locations" 
ys = [] #final label input to the network  (Steering angle)
Sd = [] #speed

T_Loss = [] #Training loss array
V_Loss = [] #Validation loss array
RMSE_loss = []  #RMSE array 

# points to the end of the last batch
train_batch_pointer = 0
val_batch_pointer = 0

####################--->> read Data <<---####################


xs = loc_frames
ys = S_angle
gs = loc_gaze 
Sd = speed 



####################--->> Train and test ratios <<---####################

train_xs = xs[:int(len(xs) * train_test_ratio)]    #training frames (portion of the whole number of frames)
train_gs = gs[:int(len(gs) * train_test_ratio)]    #training gaze maps (portion of the whole number of gaze maps)
train_ys = ys[:int(len(ys) * train_test_ratio)]    #training labels (portion of the whole number of labels)


val_xs = xs[-int(len(xs) * (1-train_test_ratio)):] #validation frames 
val_gs = gs[-int(len(gs) * (1-train_test_ratio)):] #validation gaze maps 
val_ys = ys[-int(len(ys) * (1-train_test_ratio)):] #testing labels (portion of the whole number of labels)

num_images=num_images*dataset_portion
num_train_images = len(train_gs) 
num_val_images = len(val_gs)


####################--->> LoadTrainBatch & LoadValBatch functions <<---####################

def LoadTrainBatch(batch_size):
    global train_batch_pointer
    x_out_IM = []
    x_out_GZ = []
    y_out = []
    for i in range(0, batch_size):      
        x_out_IM.append(numpy.array(Image.fromarray(matplotlib.pyplot.imread(train_xs[(train_batch_pointer + i) % num_train_images])).resize([200, 66]))/ 255.0)
        #temp = numpy.array(Image.fromarray(matplotlib.pyplot.imread(train_gs[(train_batch_pointer + i) % num_train_images])).resize([200, 66])) / 255.0
        #x_out_GZ.append(np.repeat(temp[:, :, np.newaxis], 1, axis=2))
        x_out_GZ.append(numpy.array(Image.fromarray(matplotlib.pyplot.imread(train_xs[(train_batch_pointer + i) % num_train_images])).resize([200, 66]))/ 255.0)
        y_out.append([train_ys[(train_batch_pointer + i) % num_train_images]])

    train_batch_pointer += batch_size
    return x_out_IM, x_out_GZ, y_out


def LoadValBatch(batch_size):
    global val_batch_pointer
    x_out_GZ = []
    x_out_IM = []
    y_out = []
    for i in range(0, batch_size):
        
        #temp2 = numpy.array(Image.fromarray(matplotlib.pyplot.imread(val_gs[(val_batch_pointer + i) % num_val_images])).resize([200, 66]))/ 255.0
        #x_out_GZ.append(np.repeat(temp2[:, :, np.newaxis], 1, axis=2))
        x_out_GZ.append(numpy.array(Image.fromarray(matplotlib.pyplot.imread(val_xs[(val_batch_pointer + i) % num_val_images])).resize([200, 66]))/ 255.0)
        x_out_IM.append(numpy.array(Image.fromarray(matplotlib.pyplot.imread(val_xs[(val_batch_pointer + i) % num_val_images])).resize([200, 66]))/ 255.0)
        y_out.append([val_ys[(val_batch_pointer + i) % num_val_images]])

    val_batch_pointer += batch_size
    return x_out_IM, x_out_GZ, y_out


#############################################################################################################
#--------##########################################################################################--------------#
#-----------------####################--->>> Model architicture <<<---####################---------------------------#
#--------##########################################################################################--------------#
#############################################################################################################

with tf.device('/device:GPU:0'): #to use GPU

  def weight_variable(shape):
    initial = tf.truncated_normal(shape, stddev=0.1)
    return tf.Variable(initial)

  def bias_variable(shape):
    initial = tf.constant(0.1, shape=shape)
    return tf.Variable(initial)

  def conv2d(x, W, stride):
    return tf.nn.conv2d(x, W, strides=[1, stride, stride, 1], padding='VALID')

#################################################################################################################################################
#-------------------------------------------------Inputs/output placeholder--------------------------------------------------------------------------------- 
 
  x_IMG = tf.placeholder(tf.float32, shape=[None, 66, 200, 3],name="Image")  #input frame
  x_GZ = tf.placeholder(tf.float32, shape=[None, 66, 200, 3], name="Gaze")   #input Gaze
  y_ = tf.placeholder(tf.float32, shape=[None, 1], name="Label")

  x_image = x_IMG
  x_gaze  = x_GZ
#-------------------------------------------------Image Branch---------------------------------------------------------------------------------
  #first convolutional layer
  W_conv1 = weight_variable([5, 5, 3, 24])
  b_conv1 = bias_variable([24])

  h_conv1 = tf.nn.relu(conv2d(x_image, W_conv1, 2) + b_conv1, name='IMG_1')

  grads = tf.gradients(h_conv1, x_image)

  #second convolutional layer
  W_conv2 = weight_variable([5, 5, 24, 36])
  b_conv2 = bias_variable([36])

  h_conv2 = tf.nn.relu(conv2d(h_conv1, W_conv2, 2) + b_conv2, name='IMG_2')

  #third convolutional layer
  W_conv3 = weight_variable([5, 5, 36, 48])
  b_conv3 = bias_variable([48])

  h_conv3 = tf.nn.relu(conv2d(h_conv2, W_conv3, 2) + b_conv3, name='IMG_3')

  #fourth convolutional layer
  W_conv4 = weight_variable([3, 3, 48, 64])
  b_conv4 = bias_variable([64])

  h_conv4 = tf.nn.relu(conv2d(h_conv3, W_conv4, 1) + b_conv4, name='IMG_4')

  #fifth convolutional layer
  W_conv5 = weight_variable([3, 3, 64, 36])
  b_conv5 = bias_variable([36])

  h_conv5 = tf.nn.relu(conv2d(h_conv4, W_conv5, 1) + b_conv5, name='IMG_5')
  print("OOOOOOOOOOOOOOOOOO1", h_conv5.get_shape)
#----------------------------------------------------------------Gaze Branch-------------------------------------------------------------
  #first convolutional layer
  W_conv1_gz = weight_variable([5, 5, 3, 24])
  b_conv1_gz = bias_variable([24])

  h_conv1_gz = tf.nn.relu(conv2d(x_gaze, W_conv1_gz, 2) + b_conv1_gz, name='GZ_1')
  print("First conv. shape: ",h_conv1_gz.get_shape())

  grads_gz = tf.gradients(h_conv1_gz, x_gaze)

  #second convolutional layer
  W_conv2_gz = weight_variable([5, 5, 24, 36])
  b_conv2_gz = bias_variable([36])

  h_conv2_gz = tf.nn.relu(conv2d(h_conv1_gz, W_conv2_gz, 2) + b_conv2_gz, name='GZ_2')
  print("Second conv. shape: ",h_conv2_gz.get_shape())
  #sys.exit()

  #third convolutional layer
  W_conv3_gz = weight_variable([5, 5, 36, 24]) # 48 --> 24
  b_conv3_gz = bias_variable([24])  # 48 --> 24

  h_conv3_gz = tf.nn.relu(conv2d(h_conv2_gz, W_conv3_gz, 2) + b_conv3_gz, name='GZ_3')
  print("Third conv. shape: ",h_conv3_gz.get_shape())

  #fourth convolutional layer
  #W_conv4_gz = weight_variable([3, 3, 36, 64])
  #W_conv4_gz = weight_variable([3, 3, 48, 64])
  #b_conv4_gz = bias_variable([64])

  #h_conv4_gz = tf.nn.relu(conv2d(h_conv3_gz, W_conv4_gz, 1) + b_conv4_gz)
  #print("Forth conv. shape: ",h_conv4_gz.get_shape())

  #fifth convolutional layer
  #W_conv5_gz = weight_variable([3, 3, 64, 64])
  W_conv5_gz = weight_variable([5, 5, 24, 18]) # Added # 64 --->24
  b_conv5_gz = bias_variable([18])  # 64 --->24

  h_conv5_gz = tf.nn.relu(conv2d(h_conv3_gz, W_conv5_gz, 1) + b_conv5_gz, name='GZ_4')
  print("Fifth conv. shape: ",h_conv5_gz.get_shape())
  print("OOOOOOOOOOOOOOOOOO2", h_conv5_gz.get_shape)

  #h_conv5_gz = tf.nn.relu(conv2d(h_conv2_gz, W_conv5_gz, 1) + b_conv5_gz) # Added
  #sys.exit()
#-------------------------------------------------------------------------------------------------------------------------------------
  #FCL 1

  W_fc1 = weight_variable([972, 700])
  #W_fc1 = weight_variable([2304, 1164])
  b_fc1 = bias_variable([700])

  #h_conv5_flat = tf.reshape(h_conv5, [-1, 1152])
  #h_conv5_flat_gz = tf.reshape(h_conv5_gz, [-1, 432])

  feature_vector = tf.concat(axis=-1,values=[h_conv5, h_conv5_gz])
  feature_vector = tf.reshape(feature_vector, [-1,972])

  h_fc1 = tf.nn.relu(tf.matmul(feature_vector, W_fc1) + b_fc1, name='FC1')

  keep_prob = tf.placeholder(tf.float32)
  h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)
  
  #FCL 2
  W_fc2 = weight_variable([700, 100])
  b_fc2 = bias_variable([100])

  h_fc2 = tf.nn.relu(tf.matmul(h_fc1_drop, W_fc2) + b_fc2, name='FC2')

  h_fc2_drop = tf.nn.dropout(h_fc2, keep_prob)

  #FCL 3
  W_fc3 = weight_variable([100, 50])
  b_fc3 = bias_variable([50])

  h_fc3 = tf.nn.relu(tf.matmul(h_fc2_drop, W_fc3) + b_fc3, name='FC3')

  h_fc3_drop = tf.nn.dropout(h_fc3, keep_prob)

  #FCL 3
  W_fc4 = weight_variable([50, 10])
  b_fc4 = bias_variable([10])

  h_fc4 = tf.nn.relu(tf.matmul(h_fc3_drop, W_fc4) + b_fc4, name='FC4')

  h_fc4_drop = tf.nn.dropout(h_fc4, keep_prob)

  #Output
  W_fc5 = weight_variable([10, 1])
  b_fc5 = bias_variable([1])

  #linear
  y =  tf.tanh(tf.matmul(h_fc4_drop, W_fc5) + b_fc5, name='Output')#scale the atan output

#-------------------------------------------------END OF Archtecture---------------------------------------------------------------------------------  
######################################################################################################################################################
######################################################################################################################################################
######################################################################################################################################################

#############################################################################################################
#--------##########################################################################################--------------#
#-----------------####################--->>> Training <<<---####################---------------------------#
#--------##########################################################################################--------------#
#############################################################################################################
#iteration_Number=[]
####################--->> new model training <<---####################
"""
sess = tf.InteractiveSession()
train_vars = tf.trainable_variables()
loss =tf.sqrt( tf.reduce_mean(tf.square(tf.subtract(y_,y)))) + tf.add_n([tf.nn.l2_loss(v) for v in train_vars]) * L2NormConst
print("Parameters: ",np.sum([np.product([xi.value for xi in x.get_shape()]) for x in tf.all_variables()]))
#sys.exit()
#RMSE_me = tf.sqrt(tf.reduce_mean(tf.square(tf.subtract(y_, y))))
train_step = tf.train.AdamOptimizer(learning_rate).minimize(loss)
sess.run(tf.initialize_all_variables())


####################--->> pretrained model training <<---####################
"""
sess = tf.InteractiveSession()
saver = tf.train.Saver()
saver.restore(sess, LOGDIR+"/model.ckpt")
train_vars = tf.trainable_variables()
loss =tf.sqrt( tf.reduce_mean(tf.square(tf.subtract(y_,y)))) + tf.add_n([tf.nn.l2_loss(v) for v in train_vars]) * L2NormConst
RMSE_me = tf.sqrt(tf.reduce_mean(tf.square(tf.subtract(y_, y))))
train_step = tf.train.AdamOptimizer(learning_rate).minimize(loss)
sess.run(tf.initialize_all_variables())
saver.restore(sess, LOGDIR+"/model.ckpt")


#*********************************************************************************************************************************************



# create a summary to monitor cost tensor
tf.summary.scalar("loss", loss)

# merge all summaries into a single op
merged_summary_op =  tf.summary.merge_all()

saver = tf.train.Saver(write_version = saver_pb2.SaverDef.V1)

# op to write logs to Tensorboard
logs_path = './logs'
summary_writer = tf.summary.FileWriter(logs_path, graph=tf.get_default_graph())


r=time.time()
for epoch in range(epochs):
  #print("epoch {} out of 20".format(epoch))
  t=time.time()
  for i in range(int(num_images/batch_size)):
    xs, gs, ys = LoadTrainBatch(batch_size)
    train_step.run(feed_dict={x_IMG: xs, x_GZ: gs, y_: ys, keep_prob:dropout_keep_prob})

####################--->> Modify the next loop from i%10 to i%100 after epoch 40 <<---####################


    if i % 10 == 0:
      loss_value1 = loss.eval(feed_dict={x_IMG:xs, x_GZ:gs, y_: ys, keep_prob: 1.0})         
      #xs, ys = LoadValBatch(batch_size)
      #RMSE_train = RMSE_me.eval(feed_dict={x_IMG:xs, x_GZ:gs, y_: ys, keep_prob: 1.0})
      #RMSE_train = RMSE_me.eval(feed_dict={x_IMG:xs, x_GZ:gs, y_: ys, keep_prob: 1.0})
      xs, gs, ys = LoadValBatch(batch_size)
      #xs_bk, gs_bk, ys_bk=xs, gs, ys 
      #xs2, ys2 = LoadTrainBatch(batch_size)
      #xs2, gs2, ys2 = LoadTrainBatch(batch_size)
      loss_value2 = loss.eval(feed_dict={x_IMG:xs, x_GZ:gs, y_: ys, keep_prob: 1.0})
      #RMSE_val = RMSE_me.eval(feed_dict={x_IMG:xs ,x_GZ:gs, y_: ys, keep_prob: 1.0})

      output.clear()  
      T_Loss.append(loss_value1)
      V_Loss.append(loss_value2)

      x1=time.time()-r
      hoba=(x1/((i+1)+int(num_images/batch_size)*epoch))
      x_time= ( hoba * (int(num_images/batch_size)-i ))  +  (hoba*int(num_images/batch_size)*(epochs-epoch)) #remaining time calculation
      fig = plt.figure()
      plt.plot(V_Loss, 'r')
      plt.plot(T_Loss, 'b')
      plt.legend(['V_Loss', 'T_Loss'])
      plt.show()
      '''      
      #read data.txt
      xs_new = val_xs
      gs_new = val_gs
      ys_new = val_ys
      S_angle_pridected=[]
      #get number of images
      num_images = len(gs_new)
      val=num_images
      i = num_images 
      
      ####################--->> first loop for RMSE Calculation <<---####################
      for i in range (0,num_images):
          
          image_IMG = (np.array(Image.fromarray(matplotlib.pyplot.imread(xs_new[i])).resize([200, 66]))/ 255.0)
          
          image_GZ=(np.array(Image.fromarray(matplotlib.pyplot.imread(gs_new[i])).resize([200, 66]))/ 255.0)
          image_GZ=np.repeat(image_GZ[:, :, np.newaxis], 3, axis=2)

          deg=( y.eval(feed_dict={x_IMG: [image_IMG] ,x_GZ: [image_GZ], keep_prob: 1.0})[0][0])
          S_angle_pridected.append(deg)

      RMSE =np.sqrt( np.square(np.subtract(ys_new,S_angle_pridected)).mean())

      #RMSE = np.sqrt((np.square(np.subtract(ys,S_angle_pridected))/len(ys)))
      print("*******************************************************************")
      print(" ")
      print(" ")
      print("==================>  RMSE = "+str(RMSE))
      rmse1=((0.01749223471-RMSE)/0.01749223471)*100
      print("rmse% ="+str(rmse1))
      print(" ")
      print(" ")
      print("*******************************************************************")
      '''     
      print("Epoch: %d, Step: %d, V-Loss: %g, T-Loss: %g , Remaining time: %g ,"  % (epoch, epoch * batch_size + i, loss_value2,loss_value1, (x_time)/60.0 ))


    # write logs at every iteration
    #xs, gs, ys =xs_bk, gs_bk, ys_bk
    summary = merged_summary_op.eval(feed_dict={x_IMG:xs, x_GZ:gs, y_: ys,keep_prob: 1.0})
    summary_writer.add_summary(summary, epoch * num_images/batch_size + i)
    if i % 1000 == 0:
      iteration_Number=[]
      iteration_Number = iteration_Number.append(epoch)
      with open('Epoch_Number', 'w') as f:
        ll=np.asarray(iteration_Number)
        ll.tofile(f, sep=',')
      !cp -r '/content/ll' '/content/drive/My Drive'


    if i % batch_size == 0:
      if not os.path.exists(LOGDIR):
        os.makedirs(LOGDIR)
      checkpoint_path = os.path.join(LOGDIR, "model.ckpt")
      filename = saver.save(sess, checkpoint_path)

tr_time=time.time() - r

print(" ")
print("####################--->>> Training Done <<<---####################")
print(" ")
print("total training time : "+str(tr_time/60)+" min.") # training time
print(" ")
time.sleep(3)
print("####################--->>> Starting Testing <<<---####################")
print(" ")


#saving the trained model & images
dttime=str(datetime.datetime.now())
tf.gfile.MkDir("/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Results_files/"+(name)+"_"+dttime)

checkpoint_path = os.path.join(LOGDIR, "/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Results_files/"+(name)+"_"+dttime+"/model.ckpt")
filename = saver.save(sess, checkpoint_path)

fig = plt.figure()
plt.xlabel('Number of batches')
plt.ylabel('Loss')
plt.title('Training loss and validation loss')
plt.plot(V_Loss, 'r')
plt.plot(T_Loss, 'b')

plt.legend(['V_Loss', 'T_Loss'])
plt.savefig("/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Results_files/"+(name)+"_"+dttime+"/Train_Test_Loss.png")


#############################################################################################################
#--------##########################################################################################--------------#
#-----------------####################--->>> Testing <<<---####################---------------------------#
#--------##########################################################################################--------------#
#############################################################################################################

sess = tf.InteractiveSession()
saver = tf.train.Saver()
saver.restore(sess, checkpoint_path)

#/content/drive/My Drive/Adaptive LR/ M.K_LF (exp)(Late fusion Normalizing before concatination (527) and using Tanh in output layer (removing x2) and modifying the learning_rate = 0.00005)_2020-04-10 09_52_10.735981
#read data.txt
xs = val_xs
gs = val_gs
ys = val_ys
S_angle_pridected=[]
#get number of images
num_images = len(gs)
val=num_images
i = num_images 
####################--->> first loop for RMSE Calculation <<---####################
for i in range (0,num_images):
    
    image_IMG = (np.array(Image.fromarray(matplotlib.pyplot.imread(xs[i])).resize([200, 66]))/ 255.0)
    image_GZ = (np.array(Image.fromarray(matplotlib.pyplot.imread(xs[i])).resize([200, 66]))/ 255.0)

    #image_GZ=(np.array(Image.fromarray(matplotlib.pyplot.imread(gs[i])).resize([200, 66]))/ 255.0)
    #image_GZ=np.repeat(image_GZ[:, :, np.newaxis], 1, axis=2)

    deg=( y.eval(feed_dict={x_IMG: [image_IMG] ,x_GZ: [image_GZ], keep_prob: 1.0})[0][0])
    S_angle_pridected.append(deg)

RMSE =np.sqrt( np.square(np.subtract(ys,S_angle_pridected)).mean())

print("*******************************************************************")
print(" ")
print(" ")
print("==================>  RMSE = "+str(RMSE))
rmse1=((0.010-RMSE)/0.010)*100
print("rmse% ="+str(rmse1))
print(" ")
print(" ")
print("*******************************************************************")

#read data.txt
gs = loc_frames_testing
ys = S_angle_testing
S_angle_pridected=[]

#get number of images
num_images = len(gs)
val=num_images
i = num_images 
loc_frames_testing
'''
####################--->> second loop for Visualization <<---####################

for i in range (0,num_images):

    image_IMG = (np.array(Image.fromarray(matplotlib.pyplot.imread(xs[i])).resize([200, 66]))/ 255.0)

    image_GZ = (np.array(Image.fromarray(matplotlib.pyplot.imread(gs[i])).resize([ 200,66]))/ 255.0)
    #image_GZ = np.repeat(image_GZ[:, :, np.newaxis], 3, axis=2)

    deg=( y.eval(feed_dict={x_IMG: [image_IMG] ,x_GZ: [image_GZ], keep_prob: 1.0})[0][0])
    S_angle_pridected.append(deg)


####################--->> plot the result <<---####################

fig = plt.figure(figsize=(35, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(ys, 'r')
plt.plot(S_angle_pridected, 'b')
plt.legend(['actual angle', 'predicted angle'])
plt.savefig("/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Results_files/"+(name)+"_"+dttime+"/Steering_angle.png")

'''
#############################################################################################################
#--------##########################################################################################--------------#
#-----------------####################--->>> Logging Results to DataBase <<<---####################----------------------#
#--------##########################################################################################--------------#
#############################################################################################################


####################--->> please dont modify any thing in this section ! <<---####################

if(log_flag==1):
  # Start by opening the spreadsheet and selecting the main sheet
  workbook = load_workbook(filename="/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Data_Base.xlsx")
  sheet = workbook.active

  
  flg=True 
  i=1
  while (flg):
    
    #go through the rows untile an empty one is found
    if(sheet.cell(row=i, column=1).value):
      i=i+1
      
    else:
      sheet.cell(row=i, column=1).value = str(name)
      sheet.cell(row=i, column=2).value = str( datetime.datetime.now())
      sheet.cell(row=i, column=3).value = str(RMSE)
      sheet.cell(row=i, column=4).value = str(tr_time/60)
      sheet.cell(row=i, column=5).value = str(dataset_portion)
      sheet.cell(row=i, column=6).value = str(loss_value1)
      sheet.cell(row=i, column=7).value = str(loss_value2)    
      sheet.cell(row=i, column=8).value = str(train_test_ratio)
      sheet.cell(row=i, column=9).value = str(epochs)
      sheet.cell(row=i, column=10).value = str(batch_size)
      sheet.cell(row=i, column=11).value = str(learning_rate)
      sheet.cell(row=i, column=12).value = str(L2NormConst)
      sheet.cell(row=i, column=13).value = str(dropout_keep_prob)
      sheet.cell(row=i, column=14).value = str(comment)
      #sheet.cell(row=i, column=15).value = str(rmse1)

      flg=False 

  # Save the spreadsheet
  workbook.save(filename="/content/drive/My Drive/GP/our_drEYEve/Results_DataBase/Data_Base.xlsx")

print(" ")
time.sleep(1)
print("####################--->>> Data logging Done <<<---####################")
print("**********good job "+ name+ "! **********" )
```
